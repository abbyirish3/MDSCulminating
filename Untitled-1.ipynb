### Fit bayesian model
#### without leaving out seasons

```python
# --- 1) Build X, y ---
feature_cols = ["cum_y", "t_in_season", "delta_1", "y_t"]

df_nat = df[df["region"] == "nat"].copy()
y = np.log(df_nat["peak_value"].values)  # log peak
X = df_nat[feature_cols].values

# Optional: center y to make alpha easier to sample/interpret
y_mean = y.mean()
y_centered = y - y_mean

# --- 2) Standardize X (strongly recommended) ---
scaler = StandardScaler()
Xz = scaler.fit_transform(X)

# --- 3) Fit ridge to get empirical-Bayes prior means ---

ridge = Ridge(alpha=1.0)  # you can plug in your chosen alpha
ridge.fit(Xz, y_centered)

beta_ridge = ridge.coef_.astype(float)   # shape (p,)
alpha_ridge = float(ridge.intercept_)    # should be ~0 if y_centered

# Prior scale from ridge coefficients (global)
s_beta = np.std(beta_ridge)
c = 1.5
beta_sd = max(0.1, c * s_beta)  # floor avoids overly tight priors if s_beta is tiny

# --- 4) PyMC model ---

coords = {"feature": feature_cols}

with pm.Model(coords=coords) as model:
    X_data = pm.Data("X", Xz)
    y_data = pm.Data("y", y_centered)

    # Intercept prior: centered near 0 because we centered y
    alpha = pm.Normal("alpha", mu=0.0, sigma=1.0)

    # Coefficient priors centered on ridge coefficients
    beta = pm.Normal("beta", mu=beta_ridge, sigma=beta_sd, dims="feature")

    # Residual SD prior (on log scale). Adjust if needed after prior predictive checks
    sigma = pm.HalfNormal("sigma", sigma=1.0)

    mu = alpha + pm.math.dot(X_data, beta)

    y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y_data)

    # Keep it light at first
    trace = pm.sample(
        draws=1000,
        tune=1000,
        chains=4,
        target_accept=0.9,
        random_seed=42
    )

# --- 5) Basic diagnostics ---
print(az.summary(trace, var_names=["alpha", "beta", "sigma"], round_to=3))
az.plot_trace(trace, var_names=["alpha", "beta", "sigma"]);

# --- 6) Posterior predictive distribution for peaks ---
with model:
    ppc = pm.sample_posterior_predictive(
        trace,
        var_names=["y_obs"],
        random_seed=42
    )

# ArviZ InferenceData -> posterior_predictive group
y_pred_samples = ppc.posterior_predictive["y_obs"].values  # shape (chain, draw, n)
y_pred_samples = y_pred_samples.reshape(-1, y_pred_samples.shape[-1])  # (samples, n)

# Convert back to peak scale (undo centering and log)
y_pred_samples = y_pred_samples + y_mean
peak_pred_samples = np.exp(y_pred_samples)
```

```python
median_pred = np.median(peak_pred_samples, axis=0)

lower_50 = np.percentile(peak_pred_samples, 25, axis=0)
upper_50 = np.percentile(peak_pred_samples, 75, axis=0)

lower_80 = np.percentile(peak_pred_samples, 10, axis=0)
upper_80 = np.percentile(peak_pred_samples, 90, axis=0)

lower_95 = np.percentile(peak_pred_samples, 2.5, axis=0)
upper_95 = np.percentile(peak_pred_samples, 97.5, axis=0)
```

### Leave-one-season-out Bayesian evaluation
#### pooling all weeks together in one regression for peak

```python
# -----------------------------
# CONFIG
# -----------------------------
feature_cols = ["cum_y", "t_in_season", "delta_1", "y_t"]

# National only
df_nat = df[df["region"] == "nat"].copy()

# Keep only needed rows
needed_cols = ["season", "peak_value", "t_in_season"] + feature_cols
df_nat = df_nat.dropna(subset=needed_cols).copy()

# Sort for sanity
sort_cols = ["season"]
if "epiweek" in df_nat.columns:
    sort_cols.append("epiweek")
df_nat = df_nat.sort_values(sort_cols).reset_index(drop=True)

seasons = sorted(df_nat["season"].unique())

all_results = []

# -----------------------------
# LEAVE-ONE-SEASON-OUT LOOP
# -----------------------------
for held_out in seasons:
    print(f"Running held-out season: {held_out}")

    train_df = df_nat[df_nat["season"] != held_out].copy()
    test_df = df_nat[df_nat["season"] == held_out].copy()

    # -----------------------------
    # TRAIN DATA
    # -----------------------------
    X_train = train_df[feature_cols].values
    y_train = np.log(train_df["peak_value"].values)

    # Center y using training data only
    y_mean_train = y_train.mean()
    y_train_centered = y_train - y_mean_train

    # Scale X using training data only
    scaler = StandardScaler()
    X_train_z = scaler.fit_transform(X_train)
    X_test_z = scaler.transform(test_df[feature_cols].values)

    # -----------------------------
    # RIDGE FOR EMPIRICAL-BAYES PRIOR
    # -----------------------------
    ridge = Ridge(alpha=1.0)
    ridge.fit(X_train_z, y_train_centered)

    beta_ridge = ridge.coef_.astype(float)

    s_beta = np.std(beta_ridge)
    c = 1.5
    beta_sd = max(0.1, c * s_beta)

    coords = {"feature": feature_cols}

    # -----------------------------
    # BAYESIAN MODEL
    # -----------------------------
    with pm.Model(coords=coords) as model:
        X_data = pm.Data("X", X_train_z)

        alpha = pm.Normal("alpha", mu=0.0, sigma=1.0)
        beta = pm.Normal("beta", mu=beta_ridge, sigma=beta_sd, dims="feature")
        sigma = pm.HalfNormal("sigma", sigma=1.0)

        mu = pm.Deterministic("mu", alpha + pm.math.dot(X_data, beta))

        # Training likelihood only
        y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y_train_centered)

        # Fit posterior
        trace = pm.sample(
            draws=500,
            tune=500,
            chains=2,
            target_accept=0.9,
            random_seed=42,
            progressbar=False
        )

        # Switch X to held-out season
        pm.set_data({"X": X_test_z})

        # Create prediction node AFTER fitting
        y_pred = pm.Normal("y_pred", mu=mu, sigma=sigma, shape=X_test_z.shape[0])

        # Sample posterior predictive for held-out season
        ppc = pm.sample_posterior_predictive(
            trace,
            var_names=["y_pred"],
            random_seed=42,
            progressbar=False,
            predictions=True
        )

    # -----------------------------
    # EXTRACT OUT-OF-SAMPLE PREDICTIVE SAMPLES
    # -----------------------------
    y_pred_samples = ppc.predictions["y_pred"].values   # (chains, draws, n_test)
    y_pred_samples = y_pred_samples.reshape(-1, y_pred_samples.shape[-1])  # (samples, n_test)

    # Undo centering and log-transform
    y_pred_samples = y_pred_samples + y_mean_train
    peak_pred_samples = np.exp(y_pred_samples)

    # Summaries
    pred_median = np.median(peak_pred_samples, axis=0)
    lower_80 = np.percentile(peak_pred_samples, 10, axis=0)
    upper_80 = np.percentile(peak_pred_samples, 90, axis=0)
    lower_95 = np.percentile(peak_pred_samples, 2.5, axis=0)
    upper_95 = np.percentile(peak_pred_samples, 97.5, axis=0)

    tmp = test_df.copy()
    tmp["held_out_season"] = held_out
    tmp["pred_median"] = pred_median
    tmp["lower_80"] = lower_80
    tmp["upper_80"] = upper_80
    tmp["lower_95"] = lower_95
    tmp["upper_95"] = upper_95

    all_results.append(tmp)

# Combine results
cv_results = pd.concat(all_results, ignore_index=True)

# -----------------------------
# ADD ERROR / COVERAGE METRICS
# -----------------------------
cv_results["abs_error"] = np.abs(cv_results["pred_median"] - cv_results["peak_value"])

cv_results["within_80"] = (
    (cv_results["peak_value"] >= cv_results["lower_80"]) &
    (cv_results["peak_value"] <= cv_results["upper_80"])
)

cv_results["within_95"] = (
    (cv_results["peak_value"] >= cv_results["lower_95"]) &
    (cv_results["peak_value"] <= cv_results["upper_95"])
)

cv_results["interval_width_80"] = cv_results["upper_80"] - cv_results["lower_80"]
cv_results["interval_width_95"] = cv_results["upper_95"] - cv_results["lower_95"]

print("Done.")
print(cv_results.head())
```

```python
mae_by_week = (
    cv_results.groupby("t_in_season", as_index=False)["abs_error"]
    .mean()
    .rename(columns={"abs_error": "mae"})
)

print(mae_by_week.head())
```

## separate models by forecast week

```python
# -----------------------------
# CONFIG
# -----------------------------
feature_cols = ["cum_y", "t_in_season", "delta_1", "y_t"]

# national only
df_nat = df[df["region"] == "nat"].copy()

needed_cols = ["season", "peak_value", "t_in_season"] + feature_cols
df_nat = df_nat.dropna(subset=needed_cols).copy()

# one row per season-week
df_nat = df_nat.sort_values(["season", "t_in_season"]).reset_index(drop=True)

# optionally remove very late weeks with too few seasons
week_counts = df_nat.groupby("t_in_season")["season"].nunique().reset_index(name="n_seasons")

valid_weeks = week_counts.loc[
    week_counts["n_seasons"] >= 8,
    "t_in_season"
].tolist()

# only run remaining weeks
valid_weeks = [w for w in valid_weeks if w >= 24]

all_results = []

for t in valid_weeks:
    print(f"\nRunning forecast week t = {t}")

    # -----------------------------
    # ONE ROW PER SEASON AT WEEK t
    # -----------------------------
    df_t = df_nat[df_nat["t_in_season"] == t].copy()

    seasons_t = sorted(df_t["season"].unique())

    # skip weeks with too few seasons
    if len(seasons_t) < 8:
        continue

    for held_out in seasons_t:
        train_df = df_t[df_t["season"] != held_out].copy()
        test_df  = df_t[df_t["season"] == held_out].copy()

        # should be exactly one row
        if len(test_df) != 1:
            continue

        # -----------------------------
        # TRAIN / TEST ARRAYS
        # -----------------------------
        X_train = train_df[feature_cols].values
        X_test  = test_df[feature_cols].values

        y_train = np.log(train_df["peak_value"].values)
        y_test  = np.log(test_df["peak_value"].values)

        # center y using training only
        y_mean_train = y_train.mean()
        y_train_centered = y_train - y_mean_train

        # standardize X using training only
        scaler = StandardScaler()
        X_train_z = scaler.fit_transform(X_train)
        X_test_z = scaler.transform(X_test)

        # -----------------------------
        # RIDGE PRIOR CENTER
        # -----------------------------
        ridge = Ridge(alpha=1.0)
        ridge.fit(X_train_z, y_train_centered)

        beta_ridge = ridge.coef_.astype(float)
        s_beta = np.std(beta_ridge)
        beta_sd = max(0.1, 1.5 * s_beta)

        coords = {"feature": feature_cols}

        # -----------------------------
        # BAYES MODEL FOR THIS WEEK
        # -----------------------------
        with pm.Model(coords=coords) as model:
            X_data = pm.Data("X_data", X_train_z)

            alpha = pm.Normal("alpha", mu=0.0, sigma=1.0)
            beta = pm.Normal("beta", mu=beta_ridge, sigma=beta_sd, dims="feature")
            sigma = pm.HalfNormal("sigma", sigma=1.0)

            mu = alpha + pm.math.dot(X_data, beta)

            y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y_train_centered)

            trace = pm.sample(
                draws=500,
                tune=500,
                chains=2,
                target_accept=0.95,
                random_seed=42,
                progressbar=False
            )

            # switch to held-out row
            pm.set_data({"X_data": X_test_z})

            mu_test = alpha + pm.math.dot(X_data, beta)
            y_pred = pm.Normal("y_pred", mu=mu_test, sigma=sigma)

            ppc = pm.sample_posterior_predictive(
                trace,
                var_names=["y_pred"],
                predictions=True,
                random_seed=42,
                progressbar=False
            )

        # -----------------------------
        # EXTRACT PREDICTIVE SAMPLES
        # -----------------------------
        y_pred_samples = ppc.predictions["y_pred"].values.reshape(-1)

        # un-center and back-transform
        y_pred_samples = y_pred_samples + y_mean_train
        peak_pred_samples = np.exp(y_pred_samples)

        pred_median = np.median(peak_pred_samples)
        lower_80 = np.percentile(peak_pred_samples, 10)
        upper_80 = np.percentile(peak_pred_samples, 90)
        lower_95 = np.percentile(peak_pred_samples, 2.5)
        upper_95 = np.percentile(peak_pred_samples, 97.5)

        observed_peak = float(test_df["peak_value"].iloc[0])

        all_results.append({
            "season": held_out,
            "t_in_season": t,
            "observed_peak": observed_peak,
            "pred_median": pred_median,
            "lower_80": lower_80,
            "upper_80": upper_80,
            "lower_95": lower_95,
            "upper_95": upper_95,
            "abs_error": abs(pred_median - observed_peak),
            "within_80": lower_80 <= observed_peak <= upper_80,
            "within_95": lower_95 <= observed_peak <= upper_95,
            "interval_width_80": upper_80 - lower_80,
            "interval_width_95": upper_95 - lower_95
        })

cv_weekly_results = pd.DataFrame(all_results)

output_path = "results/cv_results_weeks_24_plus.csv"
cv_weekly_results.to_csv(output_path, index=False)

print("Done.")
print(f"Saved results to {output_path}")
print(cv_weekly_results.head())
```

```python
# -----------------------------
# FIND VALID WEEKS
# -----------------------------
week_counts = (
    df_nat.groupby("t_in_season")["season"]
    .nunique()
    .reset_index(name="n_seasons")
)

valid_weeks = week_counts.loc[
    week_counts["n_seasons"] >= 8,
    "t_in_season"
].tolist()

valid_weeks = sorted(valid_weeks)

# -----------------------------
# FIT ONE NATIONAL MODEL PER WEEK
# -----------------------------
models_by_week = {}

for t in valid_weeks:
    print(f"Fitting national model for week {t}")

    df_t = df_nat[df_nat["t_in_season"] == t].copy()

    if df_t["season"].nunique() < 8:
        continue

    X_train = df_t[feature_cols].values
    y_train = np.log(df_t["peak_value"].values)

    y_mean_train = y_train.mean()
    y_train_centered = y_train - y_mean_train

    scaler = StandardScaler()
    X_train_z = scaler.fit_transform(X_train)

    ridge = Ridge(alpha=1.0)
    ridge.fit(X_train_z, y_train_centered)

    beta_ridge = ridge.coef_.astype(float)
    s_beta = np.std(beta_ridge)
    beta_sd = max(0.1, 1.5 * s_beta)

    coords = {"feature": feature_cols}

    with pm.Model(coords=coords) as model:
        alpha = pm.Normal("alpha", mu=0.0, sigma=1.0)
        beta = pm.Normal("beta", mu=beta_ridge, sigma=beta_sd, dims="feature")
        sigma = pm.HalfNormal("sigma", sigma=1.0)

        mu = alpha + pm.math.dot(X_train_z, beta)

        y_obs = pm.Normal("y_obs", mu=mu, sigma=sigma, observed=y_train_centered)

        trace = pm.sample(
            draws=500,
            tune=500,
            chains=2,
            target_accept=0.95,
            random_seed=42,
            progressbar=False
        )

    models_by_week[t] = {
        "trace": trace,
        "scaler": scaler,
        "y_mean_train": y_mean_train
    }

print("Done fitting national week-specific models.")
```
